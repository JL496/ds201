### Problem 2 ( 10 points)
Find the first quartile  (Q1), second quartile (Q2 = Median), and the third quartile(Q3) of the four numerical columns. Which variables have the lowest and highest interquartile range (IQR)? You can use the Quantile function in pandas to find these. More about quantile function is here.
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.quantile.html and here 
https://stats.stackexchange.com/questions/156778/percentile-vs-quantile-vs-quartile

```python
print(grbspec.quantile(.25))
print(grbspec.quantile(.5))
print(grbspec.quantile(.75))
iqr = grbspec.quantile(.75)-grbspec.quantile(.25)
print(iqr)
```

### Problem 4 ( 10 points)
We will find outliers in our data on this problem. For any variable, if any data falls below
the number $ Q1- 1.5 IQR$ or falls above the number $ Q3+ 1.5 IQR$ then those data points are outliers. Note that $IQR = Q3 - Q1$. Find how many outliers doeas each numerical variables have?

```python
noout = grbspec[~((grbspec < (grbspec.quantile(.25) - (1.5*iqr))) | (grbspec > (grbspec.quantile(.75) + (1.5*iqr)))).any(axis=1)]
```

```python
noout.shape
```

```python
grbspec1 = grb[['book_average_rating']].copy()
iqr1 = grbspec1.quantile(.75)-grbspec1.quantile(.25)
out1 = grbspec1[((grbspec1 < (grbspec1.quantile(.25) - (1.5*iqr1))) | (grbspec1 > (grbspec1.quantile(.75) + (1.5*iqr1)))).any(axis=1)]
print("book_average_rating outliers:", out1.shape)
print(grbspec1.head())
grbspec2 = grb[['num_ratings']].copy()
iqr2 = grbspec2.quantile(.75)-grbspec2.quantile(.25)
out2 = grbspec2[((grbspec2 < (grbspec2.quantile(.25) - (1.5*iqr2))) | (grbspec2 > (grbspec2.quantile(.75) + (1.5*iqr2)))).any(axis=1)]
print("num_ratings outliers:", out2.shape)
grbspec3 = grb[['num_reviews']].copy()
iqr3 = grbspec3.quantile(.75)-grbspec3.quantile(.25)
out3 = grbspec3[((grbspec1 < (grbspec3.quantile(.25) - (1.5*iqr3))) | (grbspec3 > (grbspec3.quantile(.75) + (1.5*iqr3)))).any(axis=1)]
print("num_reviews outliers:", out3.shape)
grbspec4 = grb[['score']].copy()
iqr4 = grbspec4.quantile(.75)-grbspec4.quantile(.25)
out4 = grbspec4[((grbspec4 < (grbspec4.quantile(.25) - (1.5*iqr4))) | (grbspec4 > (grbspec4.quantile(.75) + (1.5*iqr4)))).any(axis=1)]
print("score outliers:", out4.shape)
```
