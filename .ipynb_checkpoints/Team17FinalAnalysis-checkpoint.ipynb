```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
pd.set_option('display.max_columns', 25)
```

```python
epd = pd.read_csv('IA-04 2018 Election Participation Demographics - Sheet1.csv')
epd.head()
```

```python
epd.dtypes
```

```python
epd.isna().sum()
```

```python
epd.shape
```

```python
epd['Turnout 18-24']=epd['Turnout 18-24'].str[:-1].astype(int)
epd['Turnout 25-34']=epd['Turnout 25-34'].str[:-1].astype(int)
epd['Turnout 35-49']=epd['Turnout 35-49'].str[:-1].astype(int)
epd['Turnout 50-64']=epd['Turnout 50-64'].str[:-1].astype(int)
epd['Turnout 65 & Over']=epd['Turnout 65 & Over'].str[:-1].astype(int)
epd.dtypes
```

```python
countyvote = epd.groupby("County")["Total Voted"].sum()
countyreg = epd.groupby("County")["Total Registered"].sum()
countyturnout = countyvote/countyreg
print(countyturnout)
```

```python
countybar = countyturnout.plot(kind='bar', color=['blue','green','purple'], title='What Percent of Registered Voters Votes in 2018', figsize=(10,5))
plt.ylabel("Voter Turnout in 2018")
```

```python
countybar = epd.groupby(['County'])['Total Turnout'].mean().plot(kind='bar', color=['blue','green','purple'], title='How Many People Voted in Each County in 2018', figsize=(10,5))
plt.ylabel("Total Voted in 2018")
```

```python
countypie = epd.groupby(['County'])['Total Voted'].sum().plot(kind='pie', autopct = '%1.1f%%', colormap ='viridis', figsize=(15,15))
plt.show()
```

```python
countypie = epd.groupby(['County'])['Total Registered'].sum().plot(kind='pie', autopct = '%1.1f%%', colormap ='viridis', figsize=(15,15))
plt.show()
```

```python
demofreq = epd.groupby("Demographic")["Total Voted"].sum()
demoreg = epd.groupby("Demographic")["Total Registered"].sum()
demoturn = demofreq/demoreg
print(demoturn)
```

```python
demographicplot = epd.groupby(['Demographic'])['Total Voted'].sum().plot(kind='bar', color=['orange','green'], title='Number of Voters of Each Demographic')
plt.ylabel("Total Voted in 2018")
```

```python
demographicplot = demoturn.plot(kind='bar', color=['orange','green'], title='Number of Voters of Each Demographic')
plt.ylabel("Total Voted in 2018")
```

```python
tovfreq = epd.groupby("Type of Vote")["Total Voted"].sum()
tovreg = epd.groupby("Type of Vote")["Total Registered"].sum()
tovturn = tovfreq/tovreg
print(tovturn)
```

```python
tovplot = epd.groupby(['Type of Vote'])['Total Voted'].sum().plot(kind='bar', color=['blue','yellow','orange','green','red'], title='Number of Voters of Each Party')
plt.ylabel("Total Voted in 2018")
```

```python
tovplot = tovturn.plot(kind='bar', color=['blue','yellow','orange','green','red'], title='Number of Voters of Each Party')
plt.ylabel("Total Voted in 2018")
```

```python
countypie = epd.groupby(['Type of Vote'])['Total Voted'].sum().plot(kind='pie', autopct = '%1.1f%%', colormap ='coolwarm', figsize=(5,5))
plt.legend()
plt.show()
```

```python
countypie = epd.groupby(['Type of Vote'])['Total Registered'].sum().plot(kind='pie', autopct = '%1.1f%%', colormap ='coolwarm', figsize=(5,5))
plt.legend()
plt.show()
```

```python
x1 = epd["Total Registered"]
y1 = epd["Total Voted"]
X1 = sm.add_constant(x1)
lr = (sm.OLS(y1,X1))
lr.fit().summary()
```

y = 0.8133x - 1.0854  
$\beta_0$ = -1.0854  
$\beta_1$ = 0.8133  
$R^2$ = 0.977  
I chose voted 65 & over as the dependent variable because it makes sense that the number of people who voted over the age of 65 would be directly due to the number of people registered to vote over the age of 65. The $R^2$ value of 0.977 means 97.7% of number of people who voted over 65 can be explained by the number of people who were registered to vote over 65. 

```python
x2 = epd["Total Registered"]
y2 = epd["Total Turnout"]
X2 = sm.add_constant(x2)
lr2 = (sm.OLS(y2,X2))
lr2.fit().summary()
```

y = -0.0695x + 59.4880  
$\beta_0$ = 59.4880  
$\beta_1$ = -0.0695   
$R^2$ = 0.011  
I chose total turnout as the dependent variable because it makes sense that the total turnout would be directly due to the number of people registered to vote 18-24. The $R^2$ value of 0.011 means 1.1% of voter turnout can be explained by the number of people registered to vote 18-24. I'm guessing this has a negative correlation because those 18-24 seem to have a lower voter turnout in general than the rest of the population.
