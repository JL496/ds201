```python
import numpy as np
import numpy.matlib
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import seaborn as sns
```

```python
iris = pd.read_csv('IRIS.csv')
iris.head()
iris.isna().sum()
```

## Problem 1 [15]
Use the CART algorithm to make a classifier model for species variables. The species column has three labels. Use all the data to make the model and predictions. Report your accuracy and the confusion matrix.

```python
x=iris.drop('species', axis=1)
y=iris.species
cart = DecisionTreeClassifier()
cart = cart.fit(x, y)
pred = cart.predict(x)
acc = metrics.accuracy_score(pred, y)
print("Accuracy:", acc)
cm = metrics.confusion_matrix(pred,y)
print(cm)
```

### Problem 2 [20]
Repeat problem 1 by dividing the data into test and train data with 80, 20 rule. Report your accuracy and confusion matrix for both the testing and training data.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 0)
cart = DecisionTreeClassifier()
cart = cart.fit(x_train, y_train)
train_pred = cart.predict(x_train)
acc1 = metrics.accuracy_score(train_pred, y_train)
print("Accuracy:", acc1)
cm1 = metrics.confusion_matrix(train_pred,y_train)
print(cm1)
```

```python
test_pred = cart.predict(x_test)
acc2 = metrics.accuracy_score(test_pred, y_test)
print("Accuracy:", acc2)
cm2 = metrics.confusion_matrix(test_pred,y_test)
print(cm2)
```
