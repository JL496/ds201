### Problem 5 [20]
a) Find two variables that have a high positive correlation and find a linear regression line ( in python) between these two variables. Decide which variable you want to use as a dependent variable and explain. 

b) Do the same as in (a) with two variables that have a high negative correlation. 

```python
x1 = epd["Registered 65 & Over"]
y1 = epd["Voted 65 & Over"]
X1 = sm.add_constant(x1)
lr = (sm.OLS(y1,X1))
lr.fit().summary()
```

y = 0.8133x - 1.0854  
$\beta_0$ = -1.0854  
$\beta_1$ = 0.8133  
$R^2$ = 0.977  
I chose voted 65 & over as the dependent variable because it makes sense that the number of people who voted over the age of 65 would be directly due to the number of people registered to vote over the age of 65. The $R^2$ value of 0.977 means 97.7% of number of people who voted over 65 can be explained by the number of people who were registered to vote over 65.
