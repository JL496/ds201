# Problem 2  [15]
Filter the uc data with only the columns you found in problem 1 and the Crime column. Now split the data to testing and training following 75%, 25% rule. Use training data to train the linear model and report $R^2$ and adjusted $R^2$. How do these compare with the $R^2$ and adjusted $R^2$ from problem 1? Explain why there is a difference in these numbers if any.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x = uc[['M','Ed','Po1','U2','Ineq','Prob']]
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.25)
X_train=sm.add_constant(x_train)
mod=sm.OLS(y_train,X_train).fit()
mod.summary()
```

The model for the training set has an $R^2$ of 0.718, which is less than the $R^2$ of 0.766 for the model of the entire data set. The model for the training set also has an adjusted $R^2$ of 0.657, which is less than the $R^2$ of 0.731 for the model of the entire data set. This is a random sample of 75% of the dataframe, so this indicates it was able to come up with a better fit for the entire dataset than this random sample.
